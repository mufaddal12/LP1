```python
#Assignment 2 : minimum in a vector - online method
%%cu

#include <iostream>
#include <chrono>
#include <cuda_runtime.h>
#include <cstdlib>
#include <time.h>

using namespace std;

typedef long long int lli;

__global__ void getSumGPU(int* a, int n)
{
    int tid = blockIdx.x * blockDim.x + threadIdx.x;
    int step_size = 1;
    int number_of_threads = n;
    while (number_of_threads > 0) {
      if (tid < number_of_threads) {
        int first = tid * step_size * 2;
        int second = first + step_size;
        
        a[first] += a[second];
            
      }
      __syncthreads();
      step_size <<= 1;
      number_of_threads >>= 1;
    }
}


lli getSeconds(chrono::time_point<std::chrono::high_resolution_clock> start_time)
{
    auto end_time = chrono::high_resolution_clock::now();
    auto elapsed = chrono::duration_cast<std::chrono::nanoseconds>(end_time - start_time);
    lli time_taken = elapsed.count();
    return time_taken;
}

void init_array(int *a, int n)
{
    for (int i = 0; i < n; i++)
        a[i] = rand() % 100;
}

void print(int *a, int n)
{
    for (int i = 0; i < n; i++)
        cout << a[i] << " ";
    cout << endl;
}

lli getCudaSum(int *d_a, int n)
{
    int threads = 1024;
    int blocks = ((n + threads - 1)/2 ) / threads;

    auto start_time = chrono::high_resolution_clock::now();
    getSumGPU<<<blocks, threads>>>(d_a, n/2);  
    lli time_for_cuda = getSeconds(start_time);
 
    int *sum = new int;
    cudaMemcpy(sum, d_a, sizeof(int), cudaMemcpyDeviceToHost);
 
    cout << "CUDA: \n";
    cout<<"Sum: "<<*sum<<endl;
    cout<< "Time taken: " << time_for_cuda << endl << endl;
 
    return time_for_cuda;
}

lli getCPUSum(int *a, int n)
{
    int sum = a[0];
    auto start_time = chrono::high_resolution_clock::now();
    for (int i = 1; i < n; i++)
    {
          sum += a[i];
    }
    lli time_for_cpu = getSeconds(start_time);
    cout << "CPU: \n";
    cout << "Sum: " << sum << endl;
    cout << "Time taken: " << (time_for_cpu) << endl;
    cout << endl;
    return time_for_cpu;
}

int main()
{
    srand(time(0));
    int n = (1 << 12);
    int *a;
    int *d_a;
 
    a = new int[n];
    init_array(a, n);


    size_t size = n * sizeof(int);
    cudaMalloc((void **)&d_a, size);
    cudaMemcpy(d_a, a, size, cudaMemcpyHostToDevice);
 
    lli t_cuda = getCudaSum(d_a, n);
    lli t_cpu = getCPUSum(a, n);
 
    double ratio = t_cpu/double(t_cuda);
 
    cout<<"Faster by a ratio: "<<ratio<<endl;

    return 0;
}
```

## Standard Deviation

```python
#Assignment 2 : avg in a vector - my method
%%cu

#include <iostream>
#include <chrono>
#include <cuda_runtime.h>
#include <cstdlib>
#include <time.h>

using namespace std;

__global__ void getSumGPU(double *a, int n)
{
    int idx = blockIdx.x * blockDim.x + threadIdx.x;
    if (idx < n)
    {
        a[idx] += a[idx+n];
    }
}

__global__ void STDTermi(double *a, int n, double mean)
{
    int idx = blockIdx.x * blockDim.x + threadIdx.x;
    if (idx < n)
    {
        a[idx] -= mean;
        a[idx] *= a[idx];
    }
}

int getSeconds(chrono::time_point<std::chrono::high_resolution_clock> start_time)
{
    auto end_time = chrono::high_resolution_clock::now();
    auto elapsed = chrono::duration_cast<std::chrono::nanoseconds>(end_time - start_time);
    int time_taken = elapsed.count();
    return time_taken;
}

void init_array(double *a, int n)
{
    for (int i = 0; i < n; i++)
        a[i] = rand() % 100;
}

void print(double *a, int n)
{
    for (int i = 0; i < n; i++)
        cout << a[i] << " ";
    cout << endl;
}

int getCudaSTD(double *d_a, double *d_a1, int n, int threads)
{
    auto start_time = chrono::high_resolution_clock::now();

    int n1 = n;
    while (n1)
    {
        n1 /= 2;
        int blocks = (n1 + threads - 1) / threads;
        getSumGPU<<<blocks, threads>>>(d_a, n1);
    }

    double *mean;
    mean = new double;
    cudaMemcpy(mean, d_a, sizeof(double), cudaMemcpyDeviceToHost);
    (*mean) /= n;

    int blocks = (n + threads - 1) / threads;
    STDTermi<<<blocks, threads>>>(d_a1, n, (*mean));

    n1 = n;
    while (n1)
    {
        n1 /= 2;
        int blocks = (n1 + threads - 1) / threads;
        getSumGPU<<<blocks, threads>>>(d_a1, n1);
    }
   
    double *st_dev;
    st_dev = new double;
    cudaMemcpy(st_dev, d_a1, sizeof(double), cudaMemcpyDeviceToHost);

    *st_dev /= n;
    *st_dev = sqrt(*st_dev);

    int time_for_cuda = getSeconds(start_time);
    cout << "CUDA: \n";

    cout<< "Mean: "<<(*mean)<<endl;
    cout<< "Standard Deviation: "<<*st_dev<<endl;
    cout<< "Time taken: " << time_for_cuda<<" ns" << endl << endl;
    return time_for_cuda;
}

int getCPUSTD(double *a, int n)
{
    auto start_time = chrono::high_resolution_clock::now();
    double mean = a[0];
    for (int i = 1; i < n; i++)
    {
        mean += a[i];
    }
    mean /= n;
    double st_dev = 0;
    for(int i = 0;i<n;i++){
        a[i] -= mean;
        a[i] *= a[i];
        st_dev += a[i];
    }
    st_dev /= n;
    st_dev = sqrt(st_dev);
    int time_for_cpu = getSeconds(start_time);
    cout << "CPU: \n";
    cout << "Mean: " << mean << endl;
    cout << "Standard Deviation: "<<st_dev<<endl;
    cout << "Time taken: " << time_for_cpu<<" ns" << endl;
    cout << endl;
    return time_for_cpu;
}

int main()
{
    srand(time(0));
    int n = (1 << 18);
    int threads = 1024;
    double *a;
    double *d_a, *d_a1;
 
    a = new double[n];
    init_array(a, n);

    //print(a, n);
    size_t size = n * sizeof(double);
    cudaMalloc((void **)&d_a, size);
    cudaMemcpy(d_a, a, size, cudaMemcpyHostToDevice);
    cudaMalloc((void **)&d_a1, size);
    cudaMemcpy(d_a1, a, size, cudaMemcpyHostToDevice);
 
    int t_cuda = getCudaSTD(d_a, d_a1, n, threads);
    int t_cpu = getCPUSTD(a, n);
 
    double ratio = t_cpu/double(t_cuda);
 
    cout<<"Faster by a ratio: "<<ratio<<endl;

    return 0;
}
```

## Vector addition

```python
#  vector addition
%%cu
#include<iostream>
#include<chrono>
#include<cstdlib>
#include <chrono>
#include<time.h>

using namespace std;

__global__ void vector_add(int *out, int *a, int *b, int n) {
    int idx = blockIdx.x * blockDim.x + threadIdx.x;
    if(idx<n) out[idx] = a[idx] + b[idx];
}

void init_array(int *a, int n){
    for(int i = 1;i<=n;i++){
        a[i-1] = rand()%100;
    }
}

void print(int *a, int n)
{
    for (int i = 0; i < n; i++)
        cout << a[i] << "\t";
    cout << endl;
}

int getSeconds(chrono::time_point<std::chrono::high_resolution_clock> start_time)
{
    auto end_time = chrono::high_resolution_clock::now();
    auto elapsed = chrono::duration_cast<std::chrono::nanoseconds>(end_time - start_time);
    int time_taken = elapsed.count();
    return time_taken;
}

int main(){
    srand(time(0));

    int *a, *b, *c;
    int *d_a, *d_b, *d_c;

    int N = 1<<20;
    size_t size = N*sizeof(int);

    a = new int[N];
    b = new int[N];
    c = new int[N];

    init_array(a, N);
    init_array(b, N);

    cudaMalloc((void**)&d_a, size);
    cudaMalloc((void**)&d_b, size);
    cudaMalloc((void**)&d_c, size);

    cudaMemcpy(d_a, a, size, cudaMemcpyHostToDevice);
    cudaMemcpy(d_b, b, size, cudaMemcpyHostToDevice);

    int threads = 128;
    int blocks = (N+threads-1)/threads;

    auto start_time = chrono::high_resolution_clock::now();
    vector_add<<<blocks, threads>>>(d_c, d_a, d_b, N);
    int t_cuda = getSeconds(start_time);

    cudaMemcpy(c, d_c, size, cudaMemcpyDeviceToHost);
    start_time = chrono::high_resolution_clock::now();
    try{        
        for(int i = 0;i<N;i++){
            if(a[i] + b[i] != c[i]){
                throw -1;
            }
        }
        int t_cpu = getSeconds(start_time);
        cout<<"Answers Match!!"<<endl<<endl;   
        cout<<"A: ";print(a, N%100);
        cout<<"B: ";print(b, N%100);
        cout<<"C: ";print(c, N%100); 
        cout<<endl;
        
        cout<<"CUDA:\n";
        cout<<"Time taken: "<<t_cuda<<" ns"<<endl<<endl;
        cout<<"CPU:\n";
        cout<<"Time taken: "<<t_cpu<<" ns"<<endl<<endl;
        
        double ratio = t_cpu/double(t_cuda);
        cout<<"Faster by a ratio: "<<ratio<<endl;
    }
    catch(int){
        cout<<"Answer does not match!!";
    }

    cudaFree(d_a);
    cudaFree(d_b);
    cudaFree(d_c);
}
```

## Bubble Sort

```python
bsort_code = '''
#include <bits/stdc++.h>
#include <omp.h>
#include <chrono>

using namespace std;
typedef vector<int> vi;

void bsortSingleThread(vi &ar)
{
    int n = ar.size();
    for (int i = 0; i < n; i++)
    {
        for (int j = i; j < n - 1; j++)
        {
            if (ar[j] > ar[j + 1])
                swap(ar[j], ar[j + 1]);
        }
    }
}
void bsortMultipleThread(vi &ar)
{
    int n = ar.size();
    for (int i = 0; i < n; i++)
    {
        int f = i % 2;
#pragma omp parallel for
        for (int j = f; j < n - 1; j+=2)
        {
            if (ar[j] > ar[j + 1])
                swap(ar[j], ar[j + 1]);
        }
    }
}

int main()
{
    srand(time(0));
    int n = 1 << 15;
    vi ar1(n);
    for (int i = 0; i < n; i++)
        ar1[i] = rand() % 100;
    vi ar2 = ar1;
    auto start = chrono::steady_clock::now();
    bsortSingleThread(ar1);
    auto end = chrono::steady_clock::now();
    cout << "Time for single thread : " << chrono::duration_cast<chrono::microseconds>(end - start).count() << " microsecond" << endl;
    start = chrono::steady_clock::now();
    bsortMultipleThread(ar2);
    end = chrono::steady_clock::now();
    cout << "Time for multiple thread : " << chrono::duration_cast<chrono::microseconds>(end - start).count() << " microsecond" << endl;
    return 0;
}
'''
```

## Merge Sort

```python
msort_code='''
#include <bits/stdc++.h>
#include <omp.h>
#include <chrono>

using namespace std;
typedef vector<int> vi;

void merge(vi &ar, int low, int mid, int high)
{
    vi a(ar.begin() + low, ar.begin() + mid + 1);
    vi b(ar.begin() + mid + 1, ar.begin() + high + 1);
    int i = 0, j = 0;
    int n = a.size(), m = b.size();
    while (i < n && j < m)
    {
        if (a[i] < b[j])
        {
            ar[low++] = a[i++];
        }
        else
        {
            ar[low++] = b[j++];
        }
    }
    while (i < n)
        ar[low++] = a[i++];
    while (j < m)
        ar[low++] = b[j++];
}

void mergeSortSingleThread(vi &ar, int low, int high)
{
    if (low >= high)
        return;
    int mid = (low + high) / 2;
    mergeSortSingleThread(ar, low, mid);
    mergeSortSingleThread(ar, mid + 1, high);
    merge(ar, low, mid, high);
}

void mergeSortMultipleThread(vi &ar, int low, int high)
{
    if (low >= high)
        return;
    int mid = (low + high) / 2;
#pragma omp parallel sections
    {
#pragma omp section
        {
            mergeSortSingleThread(ar, low, mid);
        }
#pragma omp section
        {
            mergeSortSingleThread(ar, mid + 1, high);
        }
    }
    merge(ar, low, mid, high);
}

int main()
{
    srand(time(0));
    int n = 1 << 15;
    vi ar1(n);
    for (int i = 0; i < n; i++)
        ar1[i] = rand() % 100;
    vi ar2 = ar1;
    auto start = chrono::steady_clock::now();
    mergeSortSingleThread(ar1, 0, n - 1);
    auto end = chrono::steady_clock::now();
    cout << "Time for single thread : " << chrono::duration_cast<chrono::microseconds>(end - start).count() << " microsecond" << endl;
    start = chrono::steady_clock::now();
    mergeSortMultipleThread(ar2, 0, n-1);
    end = chrono::steady_clock::now();
    cout << "Time for multiple thread : " << chrono::duration_cast<chrono::microseconds>(end - start).count() << " microsecond" << endl;
    return 0;
}
'''
```

# Assignment 5 - 8 Puzzle Problem
Solve 8-puzzle problem using A* algorithm. Assume any initial configuration and define goal configuration clearly.



```python
import random
import copy
from queue import PriorityQueue
random.seed(a=1)
```

```python
dirs = [
        (-1, 0),
        (0, 1),
        (1, 0),
        (0, -1)
]
obj_id = 0
def inRange(r, c, n):
  return r>=0 and r<n and c>=0 and c<n
```

```python
class Node():
  def __init__(self, state, parent=None, depth=0):
    global obj_id
    self.state = copy.deepcopy(state)
    self.parent = parent
    self.depth = depth
    self.side = len(state)
    self.id = obj_id
    obj_id += 1

  def getNextStates(self):
    possible_states = []
    r = 0
    c = 0
    for i in range(self.side):
      for j in range(self.side):
        if self.state[i][j] == 0:
          r = i
          c = j
    for dr, dc in dirs:
      nextr = r + dr
      nextc = c + dc
      if inRange(nextr, nextc, self.side):
        self.state[r][c], self.state[nextr][nextc] = self.state[nextr][nextc], self.state[r][c]
        node = Node(self.state, self, self.depth+1)
        possible_states.append(node)
        self.state[r][c], self.state[nextr][nextc] = self.state[nextr][nextc], self.state[r][c]
    return possible_states

  def __sub__(self, other):
    diff = 0
    for i in range(self.side):
      for j in range(self.side):
        if self.state[i][j] != other.state[i][j]:
          diff += 1
    return diff

  def __eq__(self, other):
    if other is None:
      return False
    
    for i in range(self.side):
      for j in range(self.side):
        if self.state[i][j] != other.state[i][j]:
          return False
    return True

  def __str__(self):
    s = ""
    for i in self.state:
      s += str(i) + '\n'
    return s
```

```python
class Puzzle():
  def __init__(self):
    self.side = 3
    self.count = 8
    self.initial_state = self.generate_state()
    self.final_state = self.get_final_state()
    
  def generate_state(self):
    state_1d = [i for i in range(self.side**2)]
    random.Random(1).shuffle(state_1d)
    state = []
    for i in range(self.side):
      state.append(state_1d[i*self.side:(i+1)*self.side])
    return Node(state)

  def get_final_state(self):
    state_1d = [i for i in range(1, self.side**2)] + [0]
    state = []
    for i in range(self.side):
      state.append(state_1d[i*self.side:(i+1)*self.side])
    return Node(state)

  def solve(self):
    pq = PriorityQueue()
    pq.put((self.initial_state-self.final_state, self.initial_state.id, self.initial_state, 0))
    self.resultant_state = None
    cnt = 0
    while self.resultant_state == None and cnt < 1000000:
      _, _, current_state, current_depth = pq.get()
      if current_state == self.final_state:
        self.resultant_state = current_state
      next_states = current_state.getNextStates()
      # next_states.sort(key=lambda x.)
      for next_state in next_states:
        if next_state != current_state.parent:
          pq.put((current_state-self.final_state + current_depth + 1, next_state.id, next_state, current_depth+1))
      if pq.empty():
        break
      cnt += 1

  def showResult(self):
    states = []
    current_state = self.resultant_state
    while current_state is not None:
      states.append(current_state)
      current_state = current_state.parent
    states.reverse()
    for state in states:
      print(state)
```

```python
#Assignment 2 : avg in a vector - online method
%%cu

#include <iostream>
#include <chrono>
#include <cuda_runtime.h>
#include <cstdlib>
#include <time.h>

using namespace std;

__global__ void getSumGPU(int* a, int n)
{
    int tid = blockIdx.x * blockDim.x + threadIdx.x;
    int step_size = 1;
    int number_of_threads = n;
    while (number_of_threads > 0) {
      if (tid < number_of_threads) {
        int first = tid * step_size * 2;
        int second = first + step_size;
        
        a[first] += a[second];
            
      }
      __syncthreads();
      step_size <<= 1;
      number_of_threads >>= 1;
    }
}

int getSeconds(chrono::time_point<std::chrono::high_resolution_clock> start_time)
{
    auto end_time = chrono::high_resolution_clock::now();
    auto elapsed = chrono::duration_cast<std::chrono::nanoseconds>(end_time - start_time);
    int time_taken = elapsed.count();
    return time_taken;
}

void init_array(int *a, int n)
{
    for (int i = 0; i < n; i++)
        a[i] = rand() % 100;
}

int getCudaSum(int *d_a, int n)
{
    int threads = 1024;
    int blocks = ((n + threads - 1)/2 ) / threads;

    auto start_time = chrono::high_resolution_clock::now();
    getSumGPU<<<blocks, threads>>>(d_a, n/2);  
    int time_for_cuda = getSeconds(start_time);
 
    int *sum = new int;
    cudaMemcpy(sum, d_a, sizeof(int), cudaMemcpyDeviceToHost);
 
    cout << "CUDA: \n";
    cout<<"Average: "<<((*sum)/double(n))<<endl;
    cout<< "Time taken: " << time_for_cuda << endl << endl;
 
    return time_for_cuda;
}

int getCPUSum(int *a, int n)
{
    int sum = a[0];
    auto start_time = chrono::high_resolution_clock::now();
    for (int i = 1; i < n; i++)
    {
          sum += a[i];
    }
    int time_for_cpu = getSeconds(start_time);
    cout << "CPU: \n";
    cout << "Average: " << sum/double(n) << endl;
    cout << "Time taken: " << (time_for_cpu) << endl;
    cout << endl;
    return time_for_cpu;
}

int main()
{
    srand(time(0));
    int n = (1 << 12);
    int *a;
    int *d_a;
 
    a = new int[n];
    init_array(a, n);


    size_t size = n * sizeof(int);
    cudaMalloc((void **)&d_a, size);
    cudaMemcpy(d_a, a, size, cudaMemcpyHostToDevice);
 
    int t_cuda = getCudaSum(d_a, n);
    int t_cpu = getCPUSum(a, n);
 
    double ratio = t_cpu/double(t_cuda);
 
    cout<<"Faster by a ratio: "<<ratio<<endl;

    return 0;
}
```
